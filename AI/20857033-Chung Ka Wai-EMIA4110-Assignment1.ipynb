```
# turn the data in the question 2 as numpy array
q2=np.array(Q2)
```

```
# separate the label and the evidence
Y=q2[:,4]
X=q2[:,1:4]
# we add a ones column in the front of the evidence matrix X in order to undergo the multiple regression
ones=np.ones(20)
X= np.insert(X, 0, ones, axis=1)
# print out the X to check whether it works corrrectly
print(X)
```

By the closed form of multiple linear regression
β=(XᵀX)^-1 XᵀY


```
q2_beta=(np.linalg.inv(np.transpose(X).dot(X)).dot(np.transpose(X))).dot(Y)
print(q2_beta)
```

From the previous codes, we can easily observe that the positive factors are P/B ratio and ROE

Q2 part B

In here we are apply the target data set to the beta value we just get expected return

```
input=np.array([1,20.1,1.2,0.16])
expected=input.dot(q2_beta)
print(expected)

# From here, we can get a likely 12% of return rate

# testing=np.array([120,88.07,-0.96])
# B=np.array([1,2.8,3])
# print(B.dot(testing))
```

```
# turn the data set into a numpy array
q3=np.array(Q3)
print(Q3)
```

```
#seperate the evidences and the labels
train_X=q3[:15,0]
train_Y=q3[:15,1]

# adding a ones columnn in front of the X matrix to do regression
ones_1=np.ones(15)
train_one_X=np.column_stack((ones_1,train_X))

# get the validation sets and do prepocessing like the upper part
validation_X=q3[15:,0]
ones_2=np.ones(5)
validation_Y=q3[15:,1]
validation_one_X= np.column_stack((ones_2,validation_X))
```

```
currentX=train_one_X # To store the current X using for training in every degree
currenttX=validation_one_X # To store the current X using for validation in every degree
min_SSR=999999999999999 # storing the minimum SSR
min_beta=np.array([]) # storing the best fit beta
round=0
# we start from the first degree
for i in range (1,60):
  degree=i
  print("Current is number",i)
  if i>1:
    #once the degree is increasing, we modify the X for X for the training and validation
    currentX=np.column_stack((currentX,np.power(train_X,i)))
    currenttX=np.column_stack((currenttX,np.power(validation_X,i)))
  #get the current beta value in this degree
  q3current_beta=(np.linalg.inv(np.transpose(currentX).dot(currentX)).dot(np.transpose(currentX))).dot(train_Y)

  #calculate the SSR for this degree
  currentSSR=np.transpose((validation_Y-currenttX.dot(q3current_beta))).dot((validation_Y-currenttX.dot(q3current_beta)))

  print("Current SSR is",currentSSR)
  #if this degree brings improvement in the SSR, we store the result in this round
  if  currentSSR<min_SSR:
    min_SSR=currentSSR
    min_beta=q3current_beta
    round=i
    print("this is the min beta :",min_beta)
    print("with the shape of :",min_beta.shape)
    print("in degree:",i,'\n')
  else:
    print("No improvement in degree of ",i,'\n')
# give us the best fit degree
print("So in degree ",round," is the best with the beta of :", min_beta, "SSR is :", min_SSR)
```

Part B, we can just apply the beta to the inpu to get the answer, but first we have to modify the X of the input to match the degree of 4 which gives us the best fit of this polynomial regression model

```
# modify the input (76.5)
test=np.array([1,76.5,76.5**2,76.5**3,76.5**4])
expect=test.dot(min_beta)
print(expect)
```

```
#data preprocessing : turning the data into numpy
temp=np.array(Q6)
print(temp)
```

```
# Get the labels and evidence
Y_=temp[:,3]
X_=temp[:,:3]
# compute prior probability
P1_=np.count_nonzero(Y_ == 1)/np.count_nonzero(Y_)
P2_=np.count_nonzero(Y_ == 2)/np.count_nonzero(Y_)
P3_=np.count_nonzero(Y_ == 3)/np.count_nonzero(Y_)
print(" the P(C=1) = ",P1_)
print(" the P(C=2) = ",P2_)
print(" the P(C=3) = ",P3_)
```

```
# estimate the Gaussian distribution parameters


mean1=np.mean(X_[Y_==1],axis=0)

mean2=np.mean(X_[Y_==2],axis=0)

mean3=np.mean(X_[Y_==3],axis=0)
print("Mean of C=1 is ",mean1)
print("Mean of C=2 is ",mean2)
print("Mean of C=3 is ",mean3)

var1=np.var(X_[Y_==1],axis=0)
var2=np.var(X_[Y_==2],axis=0)
var3=np.var(X_[Y_==3],axis=0)
print("variance of C=1 is ",var1 )
print("variance of C=2 is ",var2 )
print("variance of C=3 is ",var3 )
```

```
#calculating the Gaussian distribution
def cal( input,mean, var):
  temp=(1/np.sqrt(2*np.pi*var))*np.exp(-(input-mean)**2/(2*var))
  return temp
```

```
#using the example in the lecture notes to ensure we get the function correct
tmean=np.array([2.067,1.133])
tvar=np.array([0.042,0.336])
tinput=np.array([2,1.5])
print(cal(tinput,tmean,tvar))
```

```
#compute the model with the dataset (-5.7,-4.3,-3.34)

input=np.array([-5.7,-4.3,-3.34])

P1=cal(input,mean1,var1)
P2=cal(input,mean2,var2)
P3=cal(input,mean3,var3)

# P(X1= -5.7|C=1)
print("P(X1= -5.7|C=1)is ",P1[0])
# P(X2= -4.3|C=1)
print("P(X2= -4.3|C=1)is ",P1[1])
# P(X3= -3.34C=1)
print("P(X3= -3.34C=1)is ",P1[2])

# P(X1= -5.7|C=2)
print("P(X1= -5.7|C=2)is ",P2[0])
# P(X2= -4.3|C=2)
print("P(X2= -4.3|C=2)is ",P2[1])
# P(X3= -3.34C=2)
print("P(X3= -3.34C=2)is ",P2[2])

# P(X1= -5.7|C=3)
print("P(X1= -5.7|C=3) is ",P3[0])
# P(X2= -4.3|C=3)
print("P(X2= -4.3|C=3) is ",P3[1])
# P(X3= -3.34C=3)
print("P(X3= -3.34C=3) is ",P3[2])
```

```
# For C=1 and X=(-5.7,-4.3,-3.34)
# P(C=1|X)=P(C=1)*P(X1|C=1)*P(X2|C=1)*P(X3|C=1)
print("P(C=1|X)=P(C=1)*P(X1|C=1)*P(X2|C=1)*P(X3|C=1) is ",P1_*np.prod(P1))
# P(C=2|X)=P(C=1)*P(X1|C=2)*P(X2|C=2)*P(X3|C=2)
print("P(C=1|X)=P(C=1)*P(X1|C=2)*P(X2|C=2)*P(X3|C=2) is ",P2_*np.prod(P2))
# P(C=3|X)=P(C=3)*P(X1|C=3)*P(X2|C=3)*P(X3|C=3)
print("P(C=3|X)=P(C=3)*P(X1|C=3)*P(X2|C=3)*P(X3|C=3) is ",P3_*np.prod(P3))

result=np.array([P1_*np.prod(P1),P2_*np.prod(P2),P3_*np.prod(P3)])
print("The most likely class is Class ",np.argmax(result)+1)
```
